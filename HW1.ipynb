```python
import requests
import StringIO as StringIO
import zipfile

uri = 'http://seanlahman.com/files/database/lahman-csv_2014-02-14.zip'

r = requests.get(uri)
```

#### Problem 1(b)

Summarize the Salaries DataFrame to show the total salaries for each team for each year. Show the head of the new summarized DataFrame. 

```python
df_sal_teams = df_salaries.groupby(['yearID', 'teamID'], as_index=False).sum()

df_sal_teams.head()
```

#### Problem 1(c)

Merge the new summarized Salaries DataFrame and Teams DataFrame together to create a new DataFrame
showing wins and total salaries for each team for each year year. Show the head of the new merged DataFrame.

**Hint**: Merge the DataFrames using `teamID` and `yearID`.

```python
df_merged = pd.merge(df_sal_teams, df_teams, how='inner', on=['teamID', 'yearID'])

df_merged[['yearID','teamID','salary','W']].head()
```

#### Problem 1(e):

**For AC209 Students**: Fit a linear regression to the data from each year and obtain the residuals. Plot the residuals against time to detect patterns that support your answer in 1(d). 

```python
import statsmodels.api as sm

y = df_merged['W']
X = df_merged['salary']
X = sm.add_constant(X)


model = sm.OLS(y, X)

results = model.fit()

print results.summary()

resid = results.resid
```

```python
plot = pd.DataFrame([resid]).T
plot.rename(columns={0:'resid'}, inplace=True )
plot['yr'] = df_merged['yearID']
plot.index = df_merged['teamID']
# plot.index.format()
# plot['salary'] = df_merged['salary']

plot.head()
```

```python
plt.plot(plot['yr'], plot['resid'], plot.index)

```

#### Problem 2(c)

Write a function to merge the `countries` and `income` data sets for any given year. 

```python
"""
Function
--------
mergeByYear

Return a merged DataFrame containing the income, 
country name and region for a given year. 

Parameters
----------
year : int
    The year of interest

Returns
-------
a DataFrame
   A pandas DataFrame with three columns titled 
   'Country', 'Region', and 'Income'. 

Example
-------
>>> mergeByYear(2010)
"""
#your code here

def mergeByYear(yr):
    
    #Pull Column corresponding to the year
    series_income = df_trans.ix[yr].values
    
    #Create a DataFrame with the Series
    data = pd.DataFrame(series_income, columns=['Income'])
    
    #assign the columns
    data['Country'] = df_trans.columns
    
    df_merged = pd.merge(left=df_countries, right=data,on=['Country'], how='inner')
    
#     df_merged.index = df_merged['Country'].values
    
#     df_merged = df_merged[df_merged.columns == yr]
                         
    return df_merged
    
    
df_new = mergeByYear(2010)

df_new.head()
```
